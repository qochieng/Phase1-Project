```python
#RESEARCH QUESTION 1: what films are currently doing the best worldwide

#get the profits data from the movies worldwide. Data is generated from Box office,data is in html
import requests

#read the Gross profit html data as csv file

url = "https://www.boxofficemojo.com/year/world/?ref_=bo_nb_cso_tab"
html = requests.get(url).content
Gross = pd.read_html(html)
df = Gross[-1]

df.to_csv('my data.csv')

df.iloc[:, 1:3]
df.head(10)
```

```python
#Bar chart to show profit by 5 top movies worldwide

ax = plt.subplots(figsize= (10,6))
bar_chart_ReleaseGroup =['The Batman','Doctor Strange ','Water Gate Bridge','Uncharted','Fantastic Beasts']
bar_chart_title ='5 Top Grossing Movies Worldwide'
bar_chart_count_label = 'Gross profit in $M'
bar_chart_series_label = 'Release Group' 

xAxis = bar_chart_ReleaseGroup
yAxis = [768,731,626,395,380]

plt.bar(xAxis,yAxis)
plt.title(bar_chart_title)
plt.xlabel(bar_chart_series_label, color= 'navy') 
plt.ylabel(bar_chart_count_label,color= 'navy' )
plt.show()
```

```python
#read the top grossing Genre films(USA&CANADA) html file as csv file
url = "https://www.boxofficemojo.com/genre/?ref_=bo_lnav_hm_shrt"
Genre = pd.read_html(url)[0]
df = Genre

df.to_csv('my data.csv')

df.head(10)
```

```python
#Bar chart shhowing the Genre against the total generated

ax = plt.subplots(figsize= (10,6))
bar_chart_Genre =['Heroine','Adventure ','Supernatural','IMAX','3D']
bar_chart_title ='TOP 5 LEADING GENRE IN USA & CANADA'
bar_chart_count_label = 'Total in $B'
bar_chart_series_label = 'Genre'

xAxis = bar_chart_Genre
yAxis = [8.5,68.9,60.7,56.6,50.7]
c =  ['red', 'black','blue','yellow','Orange']

plt.bar(xAxis,yAxis,color = c)
plt.title(bar_chart_title)
plt.xlabel(bar_chart_series_label, color = 'Red')
plt.ylabel(bar_chart_count_label, color = 'Red')
plt.show()
```

```python
#RESEARCH QUESTION 3:Which studio Brand is advisable to use based on profit

#read the top Production Studios in (USA&CANADA);convert html data as csv file

url = "https://www.boxofficemojo.com/brand/?ref_=bo_nb_gs_secondarytab"
Brand = pd.read_html(url)[0]
df = Brand

df.to_csv('my data.csv')

df.head(10)
```

```python
#Bar chart to show individual Studio performance based on profit

ax = plt.subplots(figsize= (10,6))

bar_chart_ReleaseGroup =['Marvel Comics','Legendary Pictures ','Lucasfilm','DC Comics','Pixar']
bar_chart_title ='TOP 5 PRODUCTION STUDIO IN USA & CANADA'
bar_chart_count_label = 'Gross profit in $B'
bar_chart_series_label = 'Studio Brand'

xAxis = bar_chart_ReleaseGroup
yAxis = [16.2,7.0,6.3,6.1,6.0]
c =['Blue','Black','Purple','Orange','Green']

plt.bar(xAxis,yAxis, color = c)
plt.title(bar_chart_title)
plt.xlabel(bar_chart_series_label,color = 'Blue')
plt.ylabel(bar_chart_count_label,color = 'Blue')
plt.show()
```
